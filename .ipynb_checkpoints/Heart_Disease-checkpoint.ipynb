# Data Preparation


```python
heart
```

```python
indexnames = heart[ heart['ca'] == 4].index
indexnames
```

```python
heart = heart.drop([92,158,163,164,251])
```

```python
heart
```

```python
cholindex = heart[heart['chol'] > 500].index
cholindex
```

```python
heart = heart.drop(cholindex)
```

```python
heart
```

```python
heart['thal'] = heart['thal'].replace(0 , 2)
```

```python
heart[heart['thal'] == 0].index
```

# Logistic Regression and Evaluation(Split and test)


```python
x = heart.drop('target',axis=1) #
y = heart['target'] #
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42) #split test   30% train test
```

```python
x_test
```

```python
lr = LogisticRegression()#call func
lr.fit(x_train,y_train) #fit train_data to model
```

```python
xt_predict = lr.predict(x_test) #predict data
xt_predict
```

```python
xt_pred_prob = lr.predict_proba(x_test)
xt_pred_prob
```

```python
lr.score(x_test,y_test)
```

```python
confusion_matrix(y_test, xt_predict)
```

```python
from sklearn.metrics import  classification_report as cr
cr(y_test, xt_predict)#
```

```python
print(cr(y_test, xt_predict)) #
```

```python
sns.heatmap(confusion_matrix(y_test, xt_predict).T,annot=True)
```

```python
fpr,tpr,threashold = roc_curve(y_test, xt_pred_prob[:,1])
```

```python
xt_pred_prob[:,1]# array 
```

```python
roc_auc_score(y_test, xt_pred_prob[:,1]) # Roc  y 
```

```python
plt.figure(figsize=(8,8))
plt.plot(fpr,tpr)
plt.plot(fpr,fpr)#
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```

# Naive Bayes and Evaluation(Split and test)
We use 2 type of Naive Baynes Bernoulli and Gaussian

```python
bernb = BernoulliNB()
bernb.fit(x_train, y_train)
```

```python
nb_predict = bernb.predict(x_test)
accuracy_score(y_test, nb_predict)
```

```python
bernb2 = BernoulliNB(binarize=True)
bernb2.fit(x_train, y_train)
nb_predict = bernb2.predict(x_test)
accuracy_score(y_test, nb_predict)
```
